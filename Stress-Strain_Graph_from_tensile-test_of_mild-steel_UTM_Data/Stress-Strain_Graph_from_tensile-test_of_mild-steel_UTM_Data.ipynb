# **Generate Stress vs strain graph from data (In CSV File) of tensile test of a dog bone specimen mild steel.**

```python
import matplotlib.pyplot as plt
import pandas as pd

```

```python
data = pd.read_csv("C:\\Users\\ErSin\\OneDrive\\Desktop\\mild-stee-tensiletestdatal-nitish-&-monu11.csv",names = ["Force(N)","Changeinlength(mm)","col3"])
data
```

```python
Stress = []
Force1 = data['Force(N)'].tolist()
CrossSectionArea = 33.6
for y in Force1:
    stress1 = y/CrossSectionArea
    Stress.append(stress1)

print(Stress)
```

```python
Changeinlength = data['Changeinlength(mm)'].tolist()
Changeinlength
```

```python
Strain = []
Gaugelength = 50 #in mm
for x in Changeinlength:
    strain = x/Gaugelength
    Strain.append(strain)

print(Strain)
```

```python
plt.plot(Strain,Stress , color = "red" )
plt.show()
```
